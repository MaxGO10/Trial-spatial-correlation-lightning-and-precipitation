```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#READ CSV FILES 
cor_selva_pp_cc=pd.read_csv("/content/drive/MyDrive/TRABAJO TÉCNICAS 2/DATOS T[ECNICAS 2/correlacion/correlación_selva_pp_cc.csv")
cor_selva_pp_wg=pd.read_csv("/content/drive/MyDrive/TRABAJO TÉCNICAS 2/DATOS T[ECNICAS 2/correlacion/correlación_selva_pp_wglc.csv")
cor_sierra_pp_cc=pd.read_csv("/content/drive/MyDrive/TRABAJO TÉCNICAS 2/DATOS T[ECNICAS 2/correlacion/correlación_sierra_pp_cc.csv")
cor_sierra_pp_wg=pd.read_csv("/content/drive/MyDrive/TRABAJO TÉCNICAS 2/DATOS T[ECNICAS 2/correlacion/correlación_sierra_pp_wgrc.csv")
cor_costa_pp_wg=pd.read_csv("/content/drive/MyDrive/TRABAJO TÉCNICAS 2/DATOS T[ECNICAS 2/correlacion/correlación_costa_pp_wgrc.csv")
cor_costa_pp_cc=pd.read_csv("/content/drive/MyDrive/TRABAJO TÉCNICAS 2/DATOS T[ECNICAS 2/correlacion/correlación_costa_pp_cc.csv")
```

#Create dataframe

```python
#dataframe column to array selva
cor_sel_pp_cc_a=np.array(cor_selva_pp_cc["Correlación"])
cor_sel_pp_cc_a
#dataframe column to array selva
cor_sierra_pp_cc_a=np.array(cor_sierra_pp_cc["Correlación"])
cor_sierra_pp_cc_a
#dataframe column to array costa
cor_costa_pp_cc_a=np.array(cor_costa_pp_cc["Correlación"])
cor_costa_pp_cc_a
#dataframe MONTH column to array costa
Month = np.array(cor_sierra_pp_cc["MONTH"])
```

```python
#CREATE A NEW DATAFRAME USING PREVIOUS ARRAYS
data_preproc = pd.DataFrame({
    'Month': Month, 
    'Cor. selva': cor_sel_pp_cc_a,
    'Cor. sierra': cor_sierra_pp_cc_a,
    'Cor. costa': cor_costa_pp_cc_a,})
#SAVE CSV
data_preproc.to_csv("correlación_pp_cc.csv")
```

#Plot

```python
fig, ax = plt.subplots(figsize=(7, 3), dpi = 150)
#contexto
sns.set_style("whitegrid")
sns.set_context("paper")
#definir los colores para linea
hue_colors = {"Cor. selva": "#3c6e71",
              "Cor. sierra": "#284b63",
              "Cor. costa":"#353535"}
#insertar grafica de lineas 
im = sns.lineplot(x='Month', y='value', style="variable",hue='variable',markers=True,palette=hue_colors, 
             data=pd.melt(data_preproc, ['Month']))
#definir el limite en y
ax.set_ylim([-0.1, 0.8])
ax.set_title("Pisco Precipitation (mm) vs Cloud Coverage (CC) (%) ")
ax.set_ylabel("correlation")
plt.savefig("Correlacion_pp_cc_2.png")
plt.show()
```

```python
#dataframe column to array selva
cor_sel_pp_wg_a=np.array(cor_selva_pp_wg["Correlación"])
cor_sel_pp_wg_a
#dataframe column to array sierra
cor_sierra_pp_wg_a=np.array(cor_sierra_pp_wg["Correlación"])
cor_sierra_pp_wg_a
#dataframe column to array costa
cor_costa_pp_wg_a=np.array(cor_costa_pp_wg["Correlación"])
cor_costa_pp_wg_a
#dataframe MONTH column to array costa
Month = np.array(cor_sierra_pp_cc["MONTH"])
```

```python
fig, ax = plt.subplots(figsize=(7, 3), dpi = 150)
#contexto
sns.set_style("darkgrid")
sns.set_context("paper")
#definir los colores para linea
hue_colors = {"Cor. selva": "#e63946",
              "Cor. sierra": "#457b9d",
              "Cor. costa":"#d4a373"}
#insertar grafica de lineas 
im = sns.lineplot(x='Month', y='value', 
                  hue='variable',style="variable",
                  markers=True, palette=hue_colors,
                  data=pd.melt(data_preproc2, ['Month']))
#definir el limite en y
ax.set_ylim([-0.1, 0.8])
ax.set_title("Pisco Precipitation (mm) vs  WGLC density")
ax.set_ylabel("correlation")
plt.savefig("Correlacion_pp_wg_2.png")
plt.show()
```
